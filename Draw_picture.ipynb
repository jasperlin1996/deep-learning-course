```python
%matplotlib inline
import keras
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
import numpy as np
data_x = np.arange(10).reshape(10,1)
#data_y = np.random.randn(10).reshape(10,)
data_y = np.array([1.22308782, 1.32209191, 0.82547069, 0.51323097, 1.1683462 ,
       4.10341957, 1.31960968, 2.163101  , 2.42170051, 0.60581603]).reshape(10,1)
print(data_x)
print(data_x.shape)
```

```python
def rescaleing(data):
    return (data-np.min(data))/(np.max(data)-np.min(data))

data_y = rescaleing(data_y)
plt.scatter(data_x, data_y)
plt.show()
```

```python
model = Sequential()
model.add(Dense(20, input_dim = 1, activation = 'linear'))
model.add(Dense(1, activation = 'linear'))
model.compile(optimizer = 'SGD', loss = 'mean_squared_error', metrics=['binary_crossentropy'])
```

```python
model.fit(data_x, data_y, epochs=20)
```

```python
a = np.arange(1000).reshape(1000, 1)
a = a[:, 0]/100
b = model.predict(a)
```

```python
plt.scatter(a, b)
plt.scatter(data_x, data_y)
plt.show()
```

```python
model_over = Sequential()
model_over.add(Dense(40, input_dim = 1, activation = 'relu'))
model_over.add(Dense(10, activation = 'sigmoid'))
model_over.add(Dense(1, activation = 'relu'))
model_over.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['mae'])

model_over.fit(data_x, data_y, epochs=5000)

a = np.arange(1000).reshape(1000, 1)
a = a[:, 0]/100
b = model_over.predict(a)

plt.scatter(a, b)
plt.scatter(data_x, data_y)
plt.show()
```
